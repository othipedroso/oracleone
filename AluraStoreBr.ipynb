# 3. Média de Avaliação das Lojas

```python
import pandas as pd
import matplotlib.pyplot as plt

# lendo os csv
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

# criando a lista de lojas
lojas = [loja1, loja2, loja3, loja4]

# calculando o faturamento de cada loja
for loja in lojas:
    loja['Faturamento'] = loja['Frete'] + loja['Preço']

# juntando tudo em uma base única
base = pd.concat(lojas)

# faturamento total
faturamento_total = base['Faturamento'].sum()

# faturamento por loja
faturamento_loja1 = loja1['Faturamento'].sum()
faturamento_loja2 = loja2['Faturamento'].sum()
faturamento_loja3 = loja3['Faturamento'].sum()
faturamento_loja4 = loja4['Faturamento'].sum()

# contagem de avaliações
quantidade_avaliacoes = base.groupby('Avaliação da compra').size()

# média de avaliação
media_avaliacao = base['Avaliação da compra'].mean()

# RESULTADOS
print("Faturamento total das lojas:")
print(f"Loja 1: R${faturamento_loja1:,.2f}")
print(f"Loja 2: R${faturamento_loja2:,.2f}")
print(f"Loja 3: R${faturamento_loja3:,.2f}")
print(f"Loja 4: R${faturamento_loja4:,.2f}")

print("\nQuantidade de Avaliações por Nota:")
print(quantidade_avaliacoes)

print("\nMédia de Avaliação das lojas:")
print(round(media_avaliacao, 2))
```

# 4. Produtos Mais e Menos Vendidos

```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

lojas = [loja1, loja2, loja3, loja4]
for loja in lojas:
  loja['Faturamento'] = loja['Frete'] + loja['Preço']

#juntando os csvs
base = pd.concat(lojas)

#calculando o faturamento
faturamento = base['Faturamento'].sum()

#faturamento por loja
faturamento_loja1 = loja1['Faturamento'].sum()
faturamento_loja2 = loja2['Faturamento'].sum()
faturamento_loja3 = loja3['Faturamento'].sum()
faturamento_loja4 = loja4['Faturamento'].sum()

#vpc (vendas por categoria)
vpc = base.groupby('Categoria do Produto').size()

# contagem de avaliações
quantidade_avaliacoes = base.groupby('Avaliação da compra').size()

# média da avaliação
media_avaliacao = base['Avaliação da compra'].mean()

# contando todas as vendas por produto
vendas_produto = base.groupby('Produto').size()

# ordenando as vendas ordem descrescente
vendas_ordenadas= vendas_produto.sort_values(ascending=False)

# produto mais vendido
top5 = vendas_ordenadas.head(5)

# produto menos vendido
bot5 = vendas_ordenadas.tail(5)



#resultados
print("Faturamento total das lojas:")
print("Faturamento Loja 1:")
print(faturamento_loja1)

print("Faturamento Loja 2:")
print(faturamento_loja2)

print("Faturamento Loja 3:")
print(faturamento_loja3)

print("Faturamento Loja 4:")
print(faturamento_loja4)

print("\nProdutos mais vendidos:")
print(top5)

print("\nProdutos menos vendidos:")
print(bot5)

print("Vendas por Categoria:")
print(vpc)

print("\nQuantidade de Avaliações por Nota:")
print(quantidade_avaliacoes)

print("\nMédia de Avaliação das suas lojas:")
print(round(media_avaliacao, 2))
```
